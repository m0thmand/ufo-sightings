```python
#loading the dataset
df=pd.read_csv('scrubbed.csv',
               delimiter=',', skiprows=0, low_memory=False)
```

```python
#stringify for replace
df['datetime'] = df['datetime'].astype(str)

#replace string 24:00 for proper datetime
df["datetime"] = df["datetime"].apply(lambda x: x.replace("24:00", "00:00"))

#type back to datetime
df['datetime'] = pd.to_datetime(df['datetime'])

```
